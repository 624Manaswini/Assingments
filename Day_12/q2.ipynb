```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tpot import TPOTClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv('code/data/train.csv')
test = pd.read_csv('code/data/test.csv')
```

```python
train.head()
```

```python
train = train.rename(columns={'Survived': 'class'})
```

```python
train = train.drop(columns=['Name', 'Ticket', 'Cabin'])
test = test.drop(columns=['Name', 'Ticket', 'Cabin'])
```

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
test['Embarked'] = test['Embarked'].fillna(test['Embarked'].mode()[0])
```

```python
le = LabelEncoder()
for col in ['Sex', 'Embarked']:
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
```

```python
X = train.drop(columns=['class', 'PassengerId'])
y = train['class']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python

tpot = TPOTClassifier(generations=5, population_size=20, verbosity=2)
tpot.fit(X_train, y_train)
```

```python

print("Test Accuracy:", tpot.score(X_val, y_val))
```

```python
df1 = pd.read_csv(r'code/data/pima-indians-diabetes.csv', header=None)
```

```python
df1.head()
```

```python
X = df1.iloc[:, :-1]
y = df1.iloc[:, -1]
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])

model.fit(X_train, y_train, epochs=50, batch_size=10, validation_split=0.2, verbose=1)
```

```python
loss, acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {acc}")
```
